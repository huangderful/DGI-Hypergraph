```python
from bs4 import BeautifulSoup
import json

# Assuming `html_content` contains your HTML data as a string
with open('ICD-11.html', 'r', encoding='utf-8') as file:
    html_content = file.read()
# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

def extract_categories(node):
    categories = []
    
    # Iterate over the <li> elements that represent each category or subcategory
    for li in node.find_all('li', recursive=False):
        category = {}
        
        # Extract the title and code (if present)
        title = li.find('span', class_='titleinh')
        code = li.find('span', class_='codeinh')
        
        if title:
            category['title'] = title.get_text(strip=True)
        if code:
            category['code'] = code.get_text(strip=True)
        
        # Recursively extract subcategories if they exist
        subcategory_list = li.find('ul')
        if subcategory_list:
            category['subcategories'] = extract_categories(subcategory_list)
        
        categories.append(category)
    
    return categories

# Find the main hierarchy section
main_hierarchy = soup.find_all('ul', class_='topul')

# Collect the categories
json_data = []

for hierarchy in main_hierarchy:
    categories = extract_categories(hierarchy)
    json_data.extend(categories)

# Print or save the JSON result
json_output = json.dumps(json_data, indent=4)
print(json_output)
with open('output.json', 'w', encoding='utf-8') as json_file:
    json_file.write(json_output)
```

```python
import json
from rapidfuzz import fuzz
# Load the JSON data from the output.json file
with open('output.json', 'r', encoding='utf-8') as json_file:
    data = json.load(json_file)
def find_second_highest_code(data, name):
    # Recursive function to search for the target name and return the highest score
    def search(node, target_name, best_score):
        # Score the current category or subcategory
        if 'title' in node:
            score = fuzz.token_set_ratio(node['title'].lower(), target_name.lower())
            # If this score is higher than the best found so far, update best_score
            if score > best_score[0]:
                best_score[0] = score
                best_score[3] = node["title"]
        
        # Recursively search the subcategories
        if 'subcategories' in node:
            for subcategory in node['subcategories']:
                search(subcategory, target_name, best_score)

    # Initialize the best score and corresponding code
    best_score = [0, None, "", ""]  # best_score[0] holds the score, best_score[1] holds the code
    old_best_score = 0
    # Start the search from the top-level category
    for broad_category in data[0]["subcategories"]:
        search(broad_category, name, best_score)
        if best_score[0] > old_best_score:
            old_best_score = best_score[0]
            best_score[1] = broad_category["code"]
            best_score[2] = broad_category["title"]
    return best_score

# Test the function with a common name
common_name = 'Schizophrenia'
second_highest_code = find_second_highest_code(data, common_name)

# Output the result
if second_highest_code:
    print(f"The second highest code for {common_name} is {second_highest_code}.")
else:
    print(f"{common_name} not found.")
```

```python
import pandas as pd
from tqdm import tqdm
only_common_name = pd.read_csv("./DDDB_COMMON_NAME.csv")
```
